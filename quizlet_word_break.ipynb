```python
import requests
import fitz  # PyMuPDF
from PIL import Image
from IPython.display import display

def open_and_display_google_drive_pdf(google_drive_link):
    try:
        # Extract file ID from the Google Drive link
        file_id = google_drive_link.split("/")[-2]

        # Construct the export link for downloading the file
        export_link = f"https://drive.google.com/uc?id={file_id}"

        # Download the PDF file
        response = requests.get(export_link)
        pdf_bytes = response.content

        # Open the PDF file with PyMuPDF
        pdf_document = fitz.open("pdf", pdf_bytes)

        # Iterate through each page and display it
        for page_number in range(pdf_document.page_count):
            page = pdf_document[page_number]
            pix = page.get_pixmap()
            image = Image.frombytes("RGB", (pix.width, pix.height), pix.samples)
            display(image)

        # Close the PDF document
        pdf_document.close()

    except Exception as e:
        print(f"Error: {e}")

# Replace the following link with your Google Drive link
google_drive_link = "https://drive.google.com/file/d/14opvCbJB8UK1EAwZh8zHJRAsgMB4yLqt/view?usp=sharing"
open_and_display_google_drive_pdf(google_drive_link)
```

```python
import requests
import fitz  # PyMuPDF

def read_google_drive_pdf(google_drive_link):
    try:
        # Extract file ID from the Google Drive link
        file_id = google_drive_link.split("/")[-2]

        # Construct the export link for downloading the file
        export_link = f"https://drive.google.com/uc?id={file_id}"

        # Download the PDF file
        response = requests.get(export_link)
        pdf_bytes = response.content

        # Open the PDF file with PyMuPDF
        pdf_document = fitz.open("pdf", pdf_bytes)

        # Initialize an empty string to store the text content
        text_content = ""

        # Iterate through each page and extract text
        for page_number in range(pdf_document.page_count):
            page = pdf_document[page_number]
            text_content += page.get_text()

        # Close the PDF document
        pdf_document.close()

        return text_content

    except Exception as e:
        print(f"Error: {e}")
        return None

# Replace the following link with your Google Drive link
google_drive_link = "https://drive.google.com/file/d/14opvCbJB8UK1EAwZh8zHJRAsgMB4yLqt/view?usp=sharing"
pdf_text = read_google_drive_pdf(google_drive_link)

if pdf_text:
    print(pdf_text)
```
