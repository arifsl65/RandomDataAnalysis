```python
import matplotlib.pyplot as plt

# Data
skills = ['JavaScript', 'Python', 'C', 'SQL', 'HTML', 'CSS']
percentages = [68, 70, 60, 65, 75, 65]

# Plotting the horizontal bar chart without axes
fig, ax = plt.subplots(figsize=(10, 4))
bars = ax.barh(skills, percentages, color='blue', edgecolor='black', linewidth=1.2)

# Adding skill labels and percentage labels at the beginning of the filled blue bars
for bar, skill, percentage in zip(bars, skills, percentages):
    ax.text(5, bar.get_y() + bar.get_height()/2, f'{skill}: {percentage}%', ha='left', va='center', color='white', fontsize=10)

# Remove axes
ax.axis('off')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
skills = ['JavaScript', 'Python', 'C', 'SQL', 'HTML', 'CSS']
percentages = [68, 70, 60, 65, 75, 65]

# Plotting the horizontal bar chart without axes
fig, ax = plt.subplots(figsize=(10, 4))
bars = ax.barh(skills, percentages, color='blue', edgecolor='black', linewidth=1.2)

# Adding skill labels and percentage labels at the beginning of the filled blue bars
for bar, skill, percentage in zip(bars, skills, percentages):
    ax.text(5, bar.get_y() + bar.get_height()/2, f'{skill}: {percentage}%', ha='left', va='center', color='white', fontsize=12, fontweight='bold')

# Remove axes
ax.axis('off')

# Show the plot
plt.show()
```
