```python
import pandas as pd
import yfinance as yf

def get_net_profit_to_revenue_ratio(ticker_symbol):
    try:
        # Fetch financial data for the specified ticker symbol
        company = yf.Ticker(ticker_symbol)

        # Get the latest annual financials
        financials = company.financials.loc['Net Income']

        # Select the latest annual net profit (last 4 quarters sum)
        latest_net_profit = financials.iloc[-1]

        # Get the latest annual revenue
        revenues = company.financials.loc['Total Revenue']
        latest_revenue = revenues.iloc[-1]

        # Check if net profit is at least 30% of the revenue
        if latest_net_profit / latest_revenue >= 0.15:
            return True, latest_revenue, latest_net_profit
        else:
            return False, latest_revenue, latest_net_profit
    except Exception as e:
        print(f"Error fetching data for {ticker_symbol}: {e}")
        return None, None, None

# Google Sheets URL
google_sheets_url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTtSSg31aRNUeVfPEsXvbQ1FvILpnUgaOcYidKjib2u67mIniEj2hOjnBAM3jLLOEKG-LDSnX_MSXyY/pub?output=xlsx'

# Read the Google Sheets document
try:
    # Reading specific columns ('Ticker Symbol' column)
    df = pd.read_excel(google_sheets_url, usecols=['Ticker Symbol'])

    # Convert 'Ticker Symbol' column to list
    fortune500_tickers = df['Ticker Symbol'].tolist()

    print(f"Successfully read {len(fortune500_tickers)} tickers from Google Sheets.")

    # Loop through each ticker and check the net profit to revenue ratio
    for ticker in fortune500_tickers:
        is_qualify, total_revenue, total_net_profit = get_net_profit_to_revenue_ratio(ticker)
        if is_qualify is not None:
            if is_qualify:
                print(f"Ticker: {ticker}, Total Revenue: {total_revenue}, Total Net Income: {total_net_profit}")
            else:
                print(f"{ticker} does not meet the criteria.")
        else:
            print(f"Skipping {ticker} due to an error.")

except Exception as e:
    print(f"Error reading data from Google Sheets: {e}")
    # Handle the error appropriately
```
